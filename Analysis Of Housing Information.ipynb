```python
# ullity matrix is a data-dense display which lets you quickly visually pick out patterns in data completion
import missingno as msno
%matplotlib inline
msno.matrix(train_data.sample(1460))
```

```python
# simple visualization of nullity by column:
msno.bar(train_data.sample(1460))
```

```python
# correlation heatmap measures nullity correlation: how strongly the presence or absence of one variable affects the presence of another
msno.heatmap(train_data)
```

```python
# percentage of missing values
total = train_data.isnull().sum().sort_values(ascending=False)
percent =  (train_data.isnull().sum() / train_data.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total','missing_ratio'])
missing_data
```

```python
# remove columns with 5 or more missing values
train_data_five_missing_values = train_data.drop((missing_data[missing_data['Total'] > 4]).index, 1)
train_data_five_missing_values
```

```python
#Samples (rows) with missing values are deleted from the data from which features with 5 or more missing values have been deleted.
no_zeroes_train_data = train_data_five_missing_values.dropna()
no_zeroes_train_data
```

## Question 5:
### Terminology Research

- Kurtosis - measure of how often outliers occur in a distribution.Kurtosis is              a statistical measure used to describe a characteristic of a                    dataset. When     normally distributed data is plotted on a graph,              it generally takes the form   of an upsidedown bell. This is                    called the bell curve. The plotted data that are furthest from the              mean of the data usually form the tails on each side of the curve.              Kurtosis indicates how much data resides in the tails.
- Skew - Skewness is a measurement of the distortion of symmetrical                      distribution or asymmetry in a data set. Skewness is demonstrated on a          bell curve when data points are not distributed symmetrically to the            left and right sides of the median on a bell curve. If the bell curve          is shifted to the left or the right, it is said to be skewed.

```python
# sorting data into numerical and categorical so as to check skewness and kurtosis
numerical_feats = train_data.dtypes[train_data.dtypes != "object"].index
print("Number of Numerical features: ", len(numerical_feats))

categorical_feats = train_data.dtypes[train_data.dtypes == "object"].index
print("Number of Categorical features: ", len(categorical_feats))
```

```python
# to check skewness and kurtosis we use numerical data
for col in numerical_feats:
    print('{:15}'.format(col), 
          'Skewness: {:05.2f}'.format(train_data[col].skew()) , 
          '   ' ,
          'Kurtosis: {:06.2f}'.format(train_data[col].kurt())  
         )
```

## Question 6:
### Confirmation of Distribution


```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
sns.distplot(train_data['SalePrice']);
print("Skewness: %f" % train_data['SalePrice'].skew())
print("Kurtosis: %f" % train_data['SalePrice'].kurt())
```

- As we see, the target variable SalePrice is not normally distributed.
  This can reduce the performance of the ML regression models because some       assume normal distribution 

```python
# perfoming logarithmic transformation on the objective variable
train_data['SalePrice_Log'] = np.log(train_data['SalePrice'])
sns.distplot(train_data['SalePrice_Log']);
print("Skewness: %f" % train_data['SalePrice_Log'].skew())
print("Kurtosis: %f" % train_data['SalePrice_Log'].kurt())
train_data.drop('SalePrice', axis = 1, inplace = True)
```

- The resulting distribution after logarithmic transformation looks more like     normal distribution

## Question 7:
### Check the correlation coefficient

```python
# correlation coefficient matrix heat map
corr = train_data.corr()
f ,ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corr, vmax=.8, square=True);
```

Looking at the heatmap above we can see strong correlations.The first one refers to the 'TotalBsmtSF' and '1stFlrSF' variables, and the second one refers to the 'GarageX' variables. Both cases show how significant the correlation is between these variables. Actually, this correlation is so strong that it can indicate a situation of multicollinearity. If we think about these variables, we can conclude that they give almost the same information so multicollinearity really occurs.

```python
# 10 features that are highly correlated with target variable(salesprice)
k = 10 #number of variables for heatmap
cols = corr.nlargest(k, 'SalePrice_Log')['SalePrice_Log'].index
cm = np.corrcoef(train_data[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```
